<font size="5">**Evaluation Matrix**</font>

Pendekatan analisis menggunakan machine learning untuk menentukan nasabah potensial dalam kampanye pemasaran ini dapat melibatkan penerapan model klasifikasi. Model ini akan dilatih menggunakan data pelanggan dan hasil kampanye sebelumnya untuk memprediksi kemungkinan nasabah melakukan deposito berjangka. Setelah model dilatih, evaluasi kinerja dilakukan menggunakan metrik seperti akurasi, precision, recall, dan F1-score.

Metrik evaluasi:

* True Positive (TP): Nasabah yang diprediksi melakukan deposito dan benar-benar melakukan deposito.
* False Positive (FP): Nasabah yang diprediksi melakukan deposito tetapi tidak melakukannya.
* True Negative (TN): Nasabah yang diprediksi tidak melakukan deposito dan benar-benar tidak melakukan deposito.
* False Negative (FN): Nasabah yang diprediksi tidak melakukan deposito tetapi sebenarnya melakukan deposito.

Contoh: Jika model memprediksi 100 nasabah potensial untuk deposito dan 80 dari mereka benar-benar melakukan deposito (TP), sementara 20 tidak melakukannya (FP), maka akurasi dari model dapat dihitung sebagai proporsi dari prediksi yang benar dibandingkan total prediksi. Evaluasi ini membantu bank untuk menilai seberapa efektif model dalam mengidentifikasi nasabah yang benar-benar tertarik pada produk deposito, dan mengoptimalkan strategi pemasaran mereka

* Akurasi (Accuracy)
  
    **Definisi:** Proporsi prediksi yang benar dari total prediksi.
    
    **Rumus:**
    \[ \text{Akurasi} = \frac{TP + TN}{TP + TN + FP + FN} \]
    
    **Contoh:** Jika ada 80 TP dan 10 TN dari 100 prediksi, akurasi adalah:
    \[ \text{Akurasi} = \frac{80 + 10}{100} = 0.90 \text{ atau } 90\% \]

* Precision
  
    **Definisi:** Proporsi prediksi positif yang benar dari semua prediksi positif.
    
    **Rumus:**
    \[ \text{Precision} = \frac{TP}{TP + FP} \]
    
    **Contoh:** Dengan 80 TP dan 20 FP, precision adalah:
    \[ \text{Precision} = \frac{80}{80 + 20} = 0.80 \text{ atau } 80\% \]

* Recall
  
    **Definisi:** Proporsi nasabah yang benar-benar positif yang berhasil diprediksi.
    
    **Rumus:**
    \[ \text{Recall} = \frac{TP}{TP + FN} \]
    
    **Contoh:** Dengan 80 TP dan 10 FN, recall adalah:
    \[ \text{Recall} = \frac{80}{80 + 10} = 0.89 \text{ atau } 89\% \]

* F1-Score
  
    **Definisi:** Harmonik rata-rata antara precision dan recall, menggabungkan kedua metrik tersebut.
    
    **Rumus:**
    \[ \text{F1-Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}} \]
    
    **Contoh:** Dengan precision 80% dan recall 89%, F1-Score adalah:
    \[ \text{F1-Score} = 2 \times \frac{0.80 \times 0.89}{0.80 + 0.89} \approx 0.84 \text{ atau } 84\% \]

```python
pairList = []
for i in df.select_dtypes(['object']).isna().sum().index:
    for j in df.isna().sum()[df.isna().sum() > 0].index:
        if i != j:
            grpTbl = df[[i,j]]
            grpTbl['isMissing'] = grpTbl[j].isna().apply(lambda x : 1 if  x == True else 0)
            grpTbl = grpTbl[[i, 'isMissing']].groupby(i).sum()
            print(f"{i} {j} std/avg: {grpTbl.std().item()/grpTbl.mean().item():.2f}")
            plt.bar(grpTbl.index, grpTbl.isMissing.tolist())
            pairList += [(f"{j}", f"{i}", round(grpTbl.std().item()/grpTbl.mean().item(),2))]
            plt.ylabel(f'Missing Data')
            plt.xlabel(f'{i}')
            plt.xticks(fontsize = 8, rotation = 75)
            plt.title(f'{j} - {i}')
            plt.show()
```

Dalam analisis ini, kita bertujuan untuk memvisualisasikan dan mencatat hubungan antara kolom yang memiliki nilai hilang dengan kolom kategorikal lainnya. Dengan menghitung rasio standar deviasi terhadap rata-rata (std/avg) dari jumlah data yang hilang dalam setiap kategori, kita dapat mengevaluasi seberapa besar variabilitas data yang hilang di antara kategori tersebut. Visualisasi menggunakan diagram batang menunjukkan distribusi nilai hilang dalam kategori yang berbeda, sementara pencatatan rasio std/avg membantu dalam memahami pola dan potensi keterkaitan antara kolom-kolom yang memiliki nilai hilang.

```python
srt_pairList= sorted(pairList, key = lambda x : (x[2], x[0]), reverse=True)
srt_pairList0  = []
for i in srt_pairList:
    try:
        mpdTemp = list(map(lambda x : x[0] , srt_pairList0))
        if i[0] not in mpdTemp :
            srt_pairList0  += [i]
    except:
        srt_pairList0 += [i]
srt_pairList0
```

Dalam proses ini, kita mengurutkan **`pairList`** berdasarkan rasio standar deviasi terhadap rata-rata (std/avg) secara menurun, dan kemudian menyaring daftar untuk hanya menyertakan pasangan kolom dengan nilai std tertinggi. Tujuannya adalah untuk mengidentifikasi pasangan kolom dengan ketidakpastian atau variabilitas data hilang yang paling besar, sehingga kita dapat fokus menganalisis bagian tersebut lebih mendalam. Dengan memahami area dengan variasi terbesar, kita bisa menentukan teknik penanganan data yang paling sesuai, seperti imputasi khusus atau metode lain, untuk mengatasi masalah nilai hilang dengan lebih efektif.

```python
indexMissing = []
informationDict_ = {}
dfImputed = df
for missingC, respectedC, _ in srt_pairList0:
    indexMissingTemp =  []
    print(f'{missingC} with {respectedC} and unique {df[respectedC].dropna().unique()}')
    for i in df[respectedC].dropna().unique():
        indexMissingTemp += [dfImputed[dfImputed[respectedC] == i][missingC].index]
        modeVal_ = dfImputed[dfImputed[respectedC] == i][missingC].mode()[0]
        dfImputed.loc[dfImputed[respectedC] == i, missingC] = dfImputed.loc[dfImputed[respectedC] == i, missingC].fillna(modeVal_)
        informationDict_[str(f'{missingC}-{respectedC}-{i}')] = modeVal_
    indexMissing.append(indexMissingTemp)
# indexMissing
display(dfImputed.isna().sum(), dfImputed, informationDict_)
```

Dalam analisis ini, kita mengidentifikasi kolom yang memiliki nilai hilang **`missingC`** dan kolom kategorikal yang relevan **`respectedC`**, kemudian melakukan imputasi berdasarkan nilai modus untuk setiap kategori dalam kolom **`respectedC`**. Pendekatan ini digunakan karena kita menduga data hilang mengikuti pola **Missing At Random (MAR)**, di mana probabilitas data hilang bergantung pada nilai variabel yang lain tetapi tidak pada nilai yang hilang itu sendiri. Dengan mengganti nilai hilang dengan modus dari kategori yang relevan, kita mencoba menjaga konsistensi data dalam setiap kategori dan mengurangi potensi bias akibat data yang hilang. Informasi tentang nilai modus yang digunakan untuk imputasi dicatat dalam **`informationDict_`**, dan hasil imputasi ditampilkan untuk memastikan efektivitas metode.

## 1. Logistic Regression (Logreg)

**Cara Kerja:**

Logistic Regression digunakan untuk klasifikasi biner dengan memodelkan probabilitas suatu kelas berdasarkan fitur input. Ini dilakukan dengan menghitung logit dari probabilitas menggunakan fungsi sigmoid, menghasilkan output antara 0 dan 1.

**Parameter Utama:**

- **Jenis Penalti (`penalty`)**: Metode regulasi yang diterapkan untuk mencegah overfitting, seperti 'l1' (Lasso) atau 'l2' (Ridge).
- **Kekuatan Regularisasi (`C`)**: Parameter yang mengontrol kekuatan regularisasi; nilai lebih rendah berarti regularisasi lebih kuat.

---

## 2. K-Nearest Neighbors (KNN)

**Cara Kerja:**

KNN adalah algoritma klasifikasi yang menentukan kelas suatu titik dengan mempertimbangkan kelas dari 'k' tetangga terdekatnya berdasarkan metrik jarak (misalnya, Euclidean distance).

**Parameter Utama:**

- **Jumlah Tetangga (`n_neighbors`)**: Jumlah tetangga terdekat yang digunakan untuk menentukan kelas.
- **Jenis Bobot (`weights`)**: Metode penentuan bobot untuk tetangga, seperti 'uniform' (semua tetangga memiliki bobot yang sama) atau 'distance' (bobots berdasarkan jarak).
- **Metrik Jarak (`p`)**: Parameter yang menentukan metrik jarak, di mana `p=1` untuk Manhattan distance dan `p=2` untuk Euclidean distance.

---

## 3. Decision Tree Classifier (DTC)

**Cara Kerja:**

Decision Tree membangun model keputusan berbentuk pohon dengan membuat keputusan berdasarkan fitur input yang membagi data pada setiap node. Setiap cabang mewakili keputusan yang dibuat, dan setiap daun mewakili hasil klasifikasi.

**Parameter Utama:**

- **Kriteria (`criterion`)**: Metode yang digunakan untuk menentukan kualitas split, seperti 'gini' (Gini impurity) atau 'entropy' (information gain).
- **Kedalaman Maksimum (`max_depth`)**: Kedalaman maksimum pohon untuk mencegah overfitting.
- **Jumlah Minimum Sampel (`min_samples_split`)**: Jumlah minimum sampel yang diperlukan untuk membagi node.

---

## 4. Random Forest Classifier (RF)

**Cara Kerja:**

Random Forest adalah ensemble dari banyak pohon keputusan (Decision Trees). Setiap pohon dibangun dari subset acak data dan fitur, dan keputusan akhir diambil berdasarkan voting mayoritas dari semua pohon.

**Parameter Utama:**

- **Jumlah Estimator (`n_estimators`)**: Jumlah pohon dalam hutan.
- **Kriteria (`criterion`)**: Metode untuk mengevaluasi split, seperti 'gini' atau 'entropy'.
- **Kedalaman Maksimum (`max_depth`)**: Kedalaman maksimum dari setiap pohon.

---

## 5. XGBoost Classifier (XGB)

**Cara Kerja:**

XGBoost adalah algoritma boosting yang membangun model ensemble dari pohon keputusan secara bertahap, dengan setiap pohon baru memperbaiki kesalahan dari pohon sebelumnya. Ini menggunakan gradient boosting untuk meningkatkan akurasi model.

**Parameter Utama:**

- **Jumlah Estimator (`n_estimators`)**: Jumlah pohon yang akan dibangun.
- **Laju Pembelajaran (`learning_rate`)**: Parameter yang mengontrol seberapa cepat model belajar, sering juga disebut sebagai 'eta'.
- **Kedalaman Maksimum (`max_depth`)**: Kedalaman maksimum dari setiap pohon.

```python
def grid_train(model_DF, modelDict, X_train, y_train):
    for model_name, model_info in modelDict.items():
        print(f"Performing grid search for {model_name}...")
        
        model = model_info['model']
        param_grid = model_info['param']
        
        grid_search = GridSearchCV(estimator=model,
                                   param_grid=param_grid,
                                   scoring='neg_log_loss',  
                                   cv=5,
                                   n_jobs=-1, 
                                   verbose=2)  
        
        grid_search.fit(X_train, y_train)
        bestModel = grid_search.best_estimator_
        bestModelScore = grid_search.best_score_
        
        model_DFT = pd.DataFrame({
            'model_name' : [f'BEST_{model_name}'],
            'model' : [bestModel],
            'param' : [grid_search.best_params_],
            'log loss from K-Fold' : [bestModelScore] 
        })
        model_DF = pd.concat([model_DF, model_DFT], axis = 0)
    
        
        print(f"Best parameters for {model_name}: {grid_search.best_params_}")
        print(f"Best score for {model_name}: {grid_search.best_score_}\n")
    return model_DF

model_DF = grid_train(model_DF, modelDict, X_train, y_train)
```

Secara high level, fungsi ini melakukan pencarian parameter terbaik untuk berbagai model machine learning menggunakan teknik grid search. Fungsi ini melatih setiap model dengan berbagai kombinasi parameter dan menyimpan hasil terbaik dalam DataFrame. Hasil yang disimpan mencakup model terbaik, parameter terbaik, dan skor terbaik untuk setiap model.

Secara teknis, fungsi `grid_train` menggunakan `GridSearchCV` untuk mencari kombinasi parameter terbaik berdasarkan `neg_log_loss` dengan validasi silang 5-fold. Setelah menemukan parameter terbaik untuk setiap model, informasi tersebut disimpan dalam DataFrame dan dikembalikan setelah seluruh proses selesai.

```python
def ensemble_train(model_DF, X_train, y_train):
    r_ = [(row['model_name'], row['model']) for _, row in model_DF.iterrows()]
    voting_clf = VotingClassifier(estimators=[(name, model) for name, model in r_],voting='soft')
    print(f"{r_}")
    voting_clf.fit(X_train, y_train)
    
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    f1_macro_scorer = make_scorer(f1_score, average='macro')
    cv_scores = cross_val_score(voting_clf, X_train, y_train, cv=kf, scoring='neg_log_loss')
    
    model_DFT = pd.DataFrame({
            'model_name' : [f'BEST_ensemble_models_byVoting'],
            'model' : [voting_clf],
            'param' : [voting_clf],
            'log loss from K-Fold' : [cv_scores.mean()] 
        })
    model_DF = pd.concat([model_DF, model_DFT], axis = 0)
    print(f"Ensemble model score: {cv_scores.mean():.2f}")
    return model_DF
    
model_DF = ensemble_train(model_DF, X_train, y_train)
```

Secara high level, fungsi ini membangun model ensemble menggunakan Voting Classifier yang menggabungkan beberapa model machine learning untuk meningkatkan kinerja prediksi. Model ensemble ini dilatih menggunakan data pelatihan, kemudian dievaluasi dengan validasi silang untuk mengukur kinerjanya. Hasil dari model ensemble, termasuk skor rata-rata, ditambahkan ke DataFrame yang menyimpan informasi model.

Secara teknis, fungsi `ensemble_train` pertama-tama mengumpulkan model-model terbaik dari `model_DF` dan menggabungkannya dalam Voting Classifier dengan metode voting 'soft'. Setelah melatih Voting Classifier dengan data pelatihan, fungsi ini menggunakan validasi silang 5-fold untuk menghitung `neg_log_loss`. Hasil evaluasi disimpan dalam DataFrame baru dan dicetak, kemudian DataFrame diperbarui dengan informasi model ensemble yang telah dilatih.

```python
smote_tomek = SMOTETomek(random_state=42)
X_train_resample, y_train_resample = X_train, y_train
print(f"Counter 1st {Counter(y_train_resample)}")
print(f'shape {X_train_resample.shape} {y_train_resample.shape}')
X_train_PCAresampled, y_train_PCAresampled = smote_tomek.fit_resample(X_train_resample, y_train_resample)
print(f"Counter 1st {Counter(y_train_resampled)}")
```

Langkah-langkah ini bertujuan untuk menangani ketidakseimbangan kelas dalam data pelatihan menggunakan teknik SMOTE-Tomek. Pertama, jumlah contoh dari setiap kelas dicetak sebelum resampling untuk memberikan gambaran tentang ketidakseimbangan awal. Teknik SMOTE-Tomek, yang menggabungkan metode SMOTE untuk oversampling kelas minoritas dan Tomek Links untuk menghapus contoh yang saling bertentangan, diterapkan pada data pelatihan untuk memperbaiki distribusi kelas. Setelah resampling, jumlah contoh dari setiap kelas dicetak kembali untuk menunjukkan perubahan dalam distribusi kelas dan memastikan bahwa proses resampling telah berhasil menyeimbangkan dataset.

```python
def grid_train_PCA_resample(model_DF, modelDict, X_train, y_train):
    for model_name, model_info in modelDict.items():
        print(f"Performing grid search for {model_name}...")
        
        model = model_info['model']
        param_grid = model_info['param']
        
        grid_search = GridSearchCV(estimator=model,
                                   param_grid=param_grid,
                                   scoring='neg_log_loss',  
                                   cv=5,
                                   n_jobs=-1, 
                                   verbose=2)  
        
        grid_search.fit(X_train, y_train)
        bestModel = grid_search.best_estimator_
        bestModelScore = grid_search.best_score_
        
        model_DFT = pd.DataFrame({
            'model_name' : [f'BEST_{model_name}_PCAResampled'],
            'model' : [bestModel],
            'param' : [grid_search.best_params_],
            'log loss from K-Fold' : [bestModelScore] 
        })
        model_DF = pd.concat([model_DF, model_DFT], axis = 0)
    
        
        print(f"Best parameters for {model_name}: {grid_search.best_params_}")
        print(f"Best score for {model_name}: {grid_search.best_score_}\n")
    return model_DF

model_DF = grid_train_PCA_resample(model_DF, modelDict, X_train_PCAresampled, y_train_PCAresampled)
```


Secara high level, fungsi ini melakukan pencarian grid untuk menemukan parameter terbaik bagi berbagai model machine learning yang telah diproses dengan PCA dan di-resample. Fungsi ini melatih setiap model dengan data pelatihan yang telah diproses, mencari parameter terbaik, dan menyimpan hasilnya dalam DataFrame. Hasil ini mencakup parameter terbaik dan skor terbaik dari pencarian grid untuk setiap model, yang kemudian dicetak.

Secara teknis, fungsi `grid_train_PCA_resample` melakukan pencarian grid menggunakan `GridSearchCV` untuk model-model dalam `modelDict` dengan data pelatihan yang telah diproses dengan PCA dan di-resample (`X_train`, `y_train`). Fungsi ini mengatur `GridSearchCV` dengan `neg_log_loss` sebagai metrik evaluasi dan validasi silang 5-fold. Setelah menemukan model terbaik dan parameter terbaik, informasi tersebut disimpan dalam DataFrame dan ditambahkan ke `model_DF`. Hasil pencarian, termasuk parameter terbaik dan skor terbaik, dicetak untuk setiap model sebelum DataFrame diperbarui dan dikembalikan.

```python
def ensemble_train_PCA_resample(model_DF, X_train_PCAresampled, y_train_PCAresampled):
    r_ = [(row['model_name'], row['model']) for _, row in model_DF[model_DF.model_name.str.contains('PCAResampled')].iterrows()]
    
    voting_clf = VotingClassifier(estimators=[(name, model) for name, model in r_],
                                  voting='soft')
    voting_clf.fit(X_train_PCAresampled, y_train_PCAresampled)
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    f1_macro_scorer = make_scorer(f1_score, average='macro')
    cv_scores = cross_val_score(voting_clf, X_train, y_train, cv=kf, scoring='neg_log_loss')
    
    model_DFT = pd.DataFrame({
            'model_name' : [f'BEST_ensemble_models_byVoting_PCAResampled'],
            'model' : [voting_clf],
            'param' : [voting_clf],
            'log loss from K-Fold' : [cv_scores.mean()] 
        })
    model_DF = pd.concat([model_DF, model_DFT], axis = 0)
    print(f"Ensemble model score: {cv_scores.mean():.2f}")
    return model_DF

model_DF = ensemble_train_PCA_resample(model_DF, X_train_PCAresampled, y_train_PCAresampled)
```

Secara high level, fungsi ini membangun dan mengevaluasi model ensemble menggunakan Voting Classifier yang menggabungkan model-model terbaik yang telah diproses dengan PCA dan di-resample. Fungsi ini melatih Voting Classifier dengan data pelatihan yang telah di-resample, menghitung skor evaluasi menggunakan validasi silang, dan menyimpan hasil evaluasi dalam DataFrame. Hasil skor model ensemble ditambahkan ke DataFrame yang ada dan dicetak.

Secara teknis, fungsi `ensemble_train_PCA_resample` memilih model-model terbaik dari `model_DF` yang nama modelnya mengandung 'PCAResampled'. Model-model ini digabungkan dalam Voting Classifier dengan metode voting 'soft'. Setelah melatih Voting Classifier dengan data pelatihan yang telah di-resample (`X_train_PCAresampled`, `y_train_PCAresampled`), fungsi ini mengevaluasi model menggunakan validasi silang 5-fold dengan `neg_log_loss` sebagai metrik. Skor rata-rata dari evaluasi disimpan dalam DataFrame baru, yang kemudian digabungkan dengan `model_DF` yang ada. Hasil skor model ensemble dicetak sebelum DataFrame diperbarui dan dikembalikan.

1. **Keseimbangan Precision dan Recall:**

Ini berarti F1-Score memberikan penilaian yang seimbang antara seberapa banyak prediksi positif yang benar (precision) dan seberapa banyak dari semua kasus positif yang terdeteksi dengan benar (recall). Dalam konteks ini, F1-Score sangat berguna ketika ada ketidakseimbangan antara kelas positif dan negatif, dan ketika baik precision maupun recall penting untuk keputusan model.

2. **Menangani Ketidakseimbangan Kelas:**
   - Dalam situasi di mana satu kelas jauh lebih jarang daripada yang lain (misalnya, nasabah yang melakukan deposit vs. tidak), F1-Score memberikan gambaran yang lebih baik tentang performa model pada kelas minoritas. Ini penting dalam banyak aplikasi praktis di mana penting untuk menangkap kasus positif langka dengan akurasi yang tinggi.

3. **Menghindari Kelemahan Precision dan Recall Terpisah:**
   - Precision dan recall masing-masing memiliki kekurangan. Precision hanya mengukur akurasi prediksi positif tanpa memperhitungkan berapa banyak kasus positif yang benar-benar ada. Recall, di sisi lain, hanya mengukur seberapa baik model menemukan kasus positif tanpa memperhitungkan kesalahan dalam prediksi positif. F1-Score mengatasi kelemahan ini dengan menggabungkan keduanya.

### Kenapa Juga Menyertakan Recall, Precision, dan Akurasi

1. **Recall:**
   - **Recall** penting untuk menilai seberapa banyak dari kasus positif yang sebenarnya dapat dideteksi oleh model. Ini penting dalam konteks di mana menangkap semua kasus positif (misalnya, nasabah yang melakukan deposit) sangat krusial. Model dengan recall tinggi akan lebih baik dalam menemukan semua contoh dari kelas positif.

2. **Precision:**
   - **Precision** mengukur akurasi prediksi positif. Ini penting ketika kesalahan dalam memprediksi positif memiliki dampak yang besar, seperti prediksi palsu positif yang dapat mengakibatkan biaya atau dampak negatif yang signifikan. Precision tinggi berarti model tidak hanya sering memprediksi positif, tetapi juga memprediksi dengan benar.

3. **Akurasi:**
   - **Akurasi** adalah metrik dasar yang menunjukkan proporsi prediksi yang benar (baik positif maupun negatif) dari total prediksi. Meskipun akurasi memberikan gambaran umum tentang kinerja model, ia bisa menyesatkan dalam kasus ketidakseimbangan kelas, di mana akurasi tinggi bisa dicapai dengan mengabaikan kelas minoritas. Namun, akurasi tetap penting untuk menilai performa keseluruhan model.



Menggunakan F1-Score sebagai prioritas utama memastikan bahwa model tidak hanya akurat dalam prediksi tetapi juga efektif dalam menangkap kasus positif yang penting.
